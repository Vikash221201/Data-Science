```python
import matplotlib.pyplot as plt

# Get the value counts of 'class' column
class_counts = data['class'].value_counts()

# Plot the bar graph
class_counts.plot(kind='bar')

# Add title and labels
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
data['variance'].plot(kind='hist', bins=20, edgecolor='black')

# Add title and labels
plt.title('Variance Distribution')
plt.xlabel('Variance')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
